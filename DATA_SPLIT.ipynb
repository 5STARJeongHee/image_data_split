```python
import glob
import pandas as pd
from sklearn.model_selection import train_test_split
import shutil
import os
```

```python
data_dir = 'fulldata'
file_format = 'jpg'

img_dir = ['/*.png', '/*.jpg' ]
label_format = '.xml'

seed = 1004

img_file_list = []
label_file_list = []
```

```python
for img in img_dir:
    temp = glob.glob(data_dir+img)
    for i in temp:
        img_file_list.append(i)
        temp_name = i.split('\\')
        temp_name.reverse()
        extention = temp_name[0].split('.')
        extention = '.'+extention[1]
        label_name = temp_name[0].split(extention)
        label_file_list.append(data_dir+'\\'+label_name[0]+label_format)

print(img_file_list)
print(label_file_list)
```

```python
print(len(img_file_list))
print(len(label_file_list))
```

```python
df = pd.DataFrame(
{
    'img_dir' : img_file_list,
    'label_dir': label_file_list
})
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['img_dir'], 
                                                    df['label_dir'], 
                                                    test_size=0.3, 
                                                    shuffle=True, 
                                                    random_state=seed
                                                   )
```

```python
valid_x, test_x, valid_y, test_y = train_test_split(X_test, 
                                                    y_test, 
                                                    test_size=0.3, 
                                                    shuffle=True, 
                                                    random_state=seed
                                                   )
```
